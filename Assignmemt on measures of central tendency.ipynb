## An institution to find out their student's ability in mathematics,reading and writing skills.The institution wants to do an exploratory study to check the following information.

### 1.find out how males and females participated in the test.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data=pd.read_csv('StudentsPerformance.csv')
```

```python
data
```

```python
data['gender'].value_counts()
```

### 2.What you think about the student's parental level of education?

```python
fig,(ax1,ax2,ax3)=plt.subplots(1,3,figsize=(20,10))
fig.suptitle('Comparing Parental Level of Education vs Scores for all subjects')
#math
sns.boxplot(x='parental level of education',y='math score',data=data,ax=ax1)
ax1.title.set_text('Math score vs Parents education')
ax1.set_xticklabels(ax1.get_xticklabels(),rotation=30)
#reading
sns.boxplot(x='parental level of education',y='reading score',data=data,ax=ax2)
ax2.title.set_text('Reading score vs Parents education')
ax2.set_xticklabels(ax2.get_xticklabels(),rotation=30)
#writing
sns.boxplot(x='parental level of education',y='writing score',data=data,ax=ax3)
ax3.title.set_text('Writing score vs Parents education')
ax3.set_xticklabels(ax3.get_xticklabels(),rotation=30)
```

###### observations:
1.For all the three subjects,the students who have scored the highest belonging to the parents with Masters Degree.
2.For all the three subjects,the students who have scored the lowest belong to the parents with 'some high school' degree.
3.Students whose parents have a bachelors degree have also consistently performed well.
4.The higher parental eductaion level,the better is score of that student.

### 3.Who score the most average for math,reading  and writing based on  1.Gender   2.Test performance

```python
#1.gender
data1=data.groupby('gender')[['math score','reading score','writing score']].mean()
data1
```

```python
data1.nlargest(1,['math score','reading score','writing score'])
```

###### observations
1. Males score the most average for math,reading  and writing subjects.

```python
#2.Test Preparation Course
data2=data.groupby('test preparation course')[['math score','reading score','writing score']].mean()
data2
```

```python
data2.nlargest(1,['math score','reading score','writing score'])
```

###### observations:
1.It is clear that all the students who completed the course have scored most average  in math,reading  and writing  those who did not complete

### 4.what you think about the scoring variation for math,reading and writing based on  1.Gender   2.Test preparation course


```python
fig,(ax1,ax2,ax3)=plt.subplots(1,3,figsize=(20,10))
fig.suptitle('Comparing Test Preparation Course vs Scores for all subjects')
#math
sns.boxplot(x='test preparation course',y='math score',data=data,ax=ax1)
ax1.title.set_text('Math score vs Test Preparation Course')
ax1.set_xticklabels(ax1.get_xticklabels(),rotation=30)
#reading
sns.boxplot(x='test preparation course',y='reading score',data=data,ax=ax2)
ax2.title.set_text('Reading score vs Test Preparation Course')
ax2.set_xticklabels(ax2.get_xticklabels(),rotation=30)
#writing
sns.boxplot(x='test preparation course',y='writing score',data=data,ax=ax3)
ax3.title.set_text('Writing score vs Test Preparation Course')
ax3.set_xticklabels(ax3.get_xticklabels(),rotation=30)
```

##### obsrvations:
1.It is clear that all the students who completed the course have scored better marks on an average than those who did not complete
2.The students did not complete the course are the ones who scored zero and the lowest marks.
